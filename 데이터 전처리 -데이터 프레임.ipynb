```python
# 테이블 다루기
#concat, append, join, merge, groupby, pivot_table
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format ='retina'
```

```python
#concat(행이나 열의 크기가 같을때 주로 사용)-가로: 컬럼을 늘린다, 세로- 데이터 갯수가 늘린다.
```

```python
data1={'city':['서울','부산','대구','대전','광주'],'population':[990,350,250,154,150],'temp':[13,16,14,13,15]}
df1=pd.DataFrame(data1)
df1
```

```python
#가로로 합치는 경우 #인덱스가 같은 데이터 생성
data2={'city':['서울','부산','대구','대전','광주'],'cars':[300,140,120,70,50]}
df2=pd.DataFrame(data2)
df2
```

```python
#가로로 붙이기
pd.concat((df1,df2),axis=1)
```

```python
#세로로 합치는 경우
#열(컬럼)이 같은 데이터 생성
data3={'city':['인천','울산'],
      'population':[290,120],
      'temp':[12.7,15.5]}
df3=pd.DataFrame(data3)
df3
```

```python
#기존의 인덱스를 유지하면서 합치는 경우
pd.concat([df1,df3],0)
```

```python
#기존의 인덱스 무시하고 인덱스를 새로 배정하는 경우 (ignore_index=True)
pd.concat((df1,df3),ignore_index=True)
```

```python
#세로 방향으로 샘플을 더할때는 concat 대신에 append사용 가능
df1.append(df3,ignore_index=True)
```

```python
#df1에서 city컬럼을 인덱스로 지정(set_index)
df1=pd.DataFrame(data1).set_index('city')
df1
```

```python
#인덱스와 컬럼 내용이 df1과 다른 새로운 데이터프레임 생성
data4={'city':['대전','광주','인천','울산'],
      'sale_1':[11,21,31,41],
      'sale_2':[12,22,32,42]}
df4=pd.DataFrame(data4).set_index('city')
df4
```

```python
#디폴트로 좌측 데이터프레임의 인덱스를 기준으로 붙이며 (.join)인덱스가 빈 곳에는 nan이 들어간다.
df1.join(df4)
```

```python
#위의 디폴트 동작과 같다(좌측 기준)-left
df1.join(df4,how='left')
```

```python
#좌우 데이터 프레임 인덱스 전체를 대상으로 붙이는 경우-outer(전체)
df1.join(df4,how='outer')
```

```python
#좌우 데이터프레임 인덱스 중 공통부분을 대상으로 붙이는 경우-inner(공통)
df1.join(df4,how='inner')
```

```python
#merge, 특정 열(컬럼)을 기준으로 데이터 프레임 합치기, 인덱스를 사용하지 않는다
df1=pd.DataFrame(data1)
df1
```

```python
#city를 인덱스로 사용하지 않는 데이터프레임 생성
df4=pd.DataFrame(data4)
df4
```

```python
#특정 컬럼 city를 기준으로 합친다.(how="inner")-특정
df1.merge(df4, on='city', how='inner')
```

```python
df1.merge(df4)
```

```python
#좌측 데이터프레임의 컬럼 city를 기준으로 합치는 경우(how="left")-좌측
df1.merge(df4, on='city',how='left')
```

```python
#우측 데이터프레임의 컬럼 city를 기준으로 합치는 경우
df1.merge(df4,on='city',how='right')
```

```python
#전체 데이터프레임의 컬럼 city를 기준으로 합치는 경우(how="outer")-전체
df1.merge(df4,on='city',how='outer')
```

```python
#테이블 합치기 예
#전자제품 미국 수출 데이터, logistic은 운송방법
x1={'item':['eCar','eCar','TV','TV','Refr','Refr'],
   'logistic':['ship','air','ship','air','ship','air'],
   '2019':[19,27,24,13,35,23],
   '2020':[21,102,34,87,45,30]}
dx1=pd.DataFrame(x1)
dx1['region']='US'
dx1
```

```python
#중국 수출 데이터, logistic은 운송 방법
x2={'item':['Refr','Refr','TV','TV','eCar','eCar'],
   'logistic':['ship','air','ship','air','ship','air'],
   '2019':[29,56,45,56,48,65],
   '2020':[180,42,56,89,98,79]}
dx2=pd.DataFrame(x2)
dx2['region']='China'
dx2
```

```python
#샘플 합치기, concat사용
#컬럼이 일치하으로 concat으로 샘플을 합칠 수 있다 (append도 같은 동작을 한다)
df=pd.concat([dx1,dx2],ignore_index=True)
#df=dx1.append(dx2,ignore_index=True)
df
```

```python
#필터링 -특정 조건을 만족하는 데이터를 추출하는 것 
#지역 필터링
df[df['region']=='US']
```

```python
#상품 필터링
df[df['item']=='Refr']
```

```python
#두 가지 상품 필터링
df[df['item'].isin(['TV','eCar'])]
```

```python
#두가지 조건을 만족하는 필터링
df[(df['region']=='China')&(df['item']=='Refr')]
```

```python
#참조 테이블과 합치기 -상품별 가격표를 합치느 방법, merge사용
#가격표 
price_list={'item':['Refr','eCar','TV'],
           'price':[200,1000,100]}
df_price=pd.DataFrame(price_list)
df_price
```

```python
#가격표를 반영한 (합친)데이터프레임
#item컬럼을 기준으로 합친다
df_all=df.merge(df_price,on="item")
df_all
```

```python
#새로운 컬럼 추가하기 (예: 상품별 매출액)
df_all['2020_sale']=df_all['2020']*df_all['price']
df_all
```

```python
#특정 조건 값 계산하기
#2020년 미국 수출 총액
df[df.region=='US']['2020'].sum()
```

```python
#2020년 항공 수출 평균액
df[df.logistic=='air']['2020'].mean()
```

```python
#2020년 전기차, 항공 수출 건수
df[(df.item=='eCar')&(df.logistic=='air')]['2020'].count()
```

```python
#groupby -특정조건에 맞는 그룹을 테이블로 새로 만들고, 다양하게 원하는 조작/계산을 한다
df.groupby('item')['2020'].sum().to_frame()
#품뭅결 2020 매출 합계 보기
df.groupby('item')['2020'].sum().to_frame()
```

```python
df.groupby('item')[['2019','2020']].agg(('sum','size','mean'))
```

```python
#품묵별, 지역별로 2020년 판매 평균치 보기
df.groupby(['item','region'])['2020'].mean().to_frame()
```

```python
#품묵별,지역별, 운송별로 2019,2020년 판매 합계 보기
df.groupby(['item','region','logistic'])[['2019','2020']].sum()
```

```python
#조건을 맞는 그룹을 각각 얻기
#groupby()의 실행 결과는 그룹명과 데이터프레임 두 가지를 리턴한다.
#특정 그룹만 얻으려면 get_group()를 사용한다
#item을 기준으로 그룹을 나누고 그중 TV에 해당하는 내용만 별도의 데이터프레임으로 얻는다
df.groupby(['item']).get_group('eCar')
```

```python
#item 기준으로 나눈 그룹을 각각 별도의 csv파일로 저장하는 예
for item, group in df.groupby('item'):
    
    group.to_csv(item+'.csv')
```

```python
#조건에 맞는 그룹 찾기
#품묵별 2020년 판매 평균값
df.groupby('item')['2020'].mean().to_frame()
```

```python
#2020년 품묵별 평균이 70을 넘는 "상품"들의 샘플을 모두 찾기(여기서는 eCar과 Refr 두 상품만)
#filter를 사용하면 편리하다
#filter 수행 결과는 True/False를 얻으며 해당 샘플을 추출한다
df.groupby('item').filter(
    lambda x: x['2020'].mean()>70)
#품묵(item)별로 매출이 70보다 넘는 것들
```

```python
#2020에서 70넘는 것만
df[df["2020"]>70]
```

```python
#상품별 2020년 총 매출액 #aggfunc='sum'
df.pivot_table(index=['item'],values=['2020'],aggfunc='sum')
```

```python
#상품, 지역별 2020년 매출 합계와 평균 보기
df.pivot_table(index=['item','region'],values=['2020'],aggfunc=['sum','mean'])
```

```python
#columns를 지정하면 칼럼을 세분화하여 보여준다
#위의 결과를 운송 수단별로 더 세분화 하여 보기
df.pivot_table(index=['item','region'],values=['2020'],aggfunc=['sum','mean'],columns=['logistic'])
```
